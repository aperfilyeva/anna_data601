One bar graph has 24 bars; one bar graph has 24*(number of days) bars. Submit the .ipynb file containing the analysis and the generated pictures

```python
import plotly.express as px
import plotly
import pandas as pd

data_url = 'https://raw.githubusercontent.com/aperfilyeva/anna_data601/main/assignment_05/RollingSystemDemand_20220330_1729.csv'
```

```python
df = pd.read_csv(data_url,
                         index_col=False,
                         skiprows=1,
                         skipfooter=1,
                         engine='python',
                         header=None)

df.head()
```

```python
df.columns=['VD','date_time','demand']

df.head()
```

```python
df['date_time']=pd.to_datetime(df['date_time'].astype(str))
df['date'] = df['date_time'].dt.date
df['hour'] = df['date_time'].dt.hour
df['dh']=df['date'].astype(str)+"-"+df['hour'].astype(str)
df.sample(5)
```

```python
fig = px.bar(data_frame=df, 
             x='hour', y="demand",
             labels={"demand":"Demand","hour":"Hour of Day"},
             title="Power Demand by Hour of Day",
             template=list(plotly.io.templates.keys())[5]
)
fig.update_layout(showlegend=False)

fig.show()
```

```python
fig = px.bar(data_frame=df, 
             x="dh", y="demand",
             labels={"demand":"Demand","dh":"Date-Hour"},
             title="Power Demand by Date and Hour",
             template=list(plotly.io.templates.keys())[5]
)
fig.update_layout(showlegend=False)

fig.show()
```
